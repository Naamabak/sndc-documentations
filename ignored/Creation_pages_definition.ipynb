```python
import pandas as pd
import unicodedata
import json
from pprint import pprint
from difflib import get_close_matches
```

```python
def replace_accents(any_str):
    return ''.join(c for c in unicodedata.normalize('NFD', any_str)
                   if unicodedata.category(c) != 'Mn')
```

```python
concept_rattache_map = {
    'Organisme de liquidation du bénéficiaire': 'Organisme de liquidation de la prestation'
}
```

```python
for concept in set(all_concepts_rataches) - set(all_concepts):
    if not concept:
        continue
    concept = concept_rattache_map.get(concept) or concept
    concept_proche = get_close_matches(concept, all_concepts, n=1, cutoff=0.7)
    if concept_proche:
        if concept.lower()  == concept_proche[0].lower():
            continue
        print(concept)
        print(concept_proche[0])
        print('---')
    else:
        print('##')
        print(concept)
        print('##')
```

```python
def concept_to_file_name(concept):
    return ('{}.md'.format(replace_accents(concept))
                 .lower()
                 .replace(' ', '_')
                 .replace(" ", "")
                 .replace("'","_")
                 .replace("\n", "")
                 .replace("-", '_')
                 .replace(",", '_')
                 .replace("(", '_')
                 .replace(")", '_')
                 .replace('__', '_').replace('__', '_')
                 .replace('_.', '.')
                )

concept_to_file_name("Bénéficiaire")
```

```python
concept_rattache_to_drop= ['Catégorie', 'Modalité de catégorie', 'Structure', 'Pathologie,', 'Patient']

def format_concepts_rattaches(concept_list):
    concept_str_list = []
    for concept in concept_list.split(', '):
        concept = concept_rattache_map.get(concept) or concept
        if concept in concept_rattache_to_drop:
            continue
        if not concept in all_concepts:
            concept = get_close_matches(concept, all_concepts, n=1, cutoff=0.7)[0]
        concept = concept.strip(',')
        concept_str_list.append('[{}]({})'.format(concept, concept_to_file_name(concept)))
    return ', '.join(concept_str_list)
    
     

s = "Bénéficiaire, Ayant-droit"
format_concepts_rattaches(s)
```

```python
template = """# {Concept} {Sigle}
<!-- SPDX-License-Identifier: MPL-2.0 -->

{Synonyme}Notion plus générale : {Famille}

## Définitions

{Définition}

## Remarques

{Remarques}


{rattache}

"""
def line_to_md(s):
    result = template.format(
        Concept=s.Concept,
        Sigle='({})'.format(s.Sigle) if s.Sigle else "", 
        Synonyme='Synonmyes: {}\n'.format(s.Synonyme) if s.Synonyme else "",
        Famille=s.Famille,
        Définition=s.Définition,
        Remarques=s.Remarques,
        # acces_sniiram=s["Accès aux données dans le SNIIR-AM"],
        rattache='## Notions liées\n\n{}'.format(format_concepts_rattaches(s["Concept rattaché"])) if s["Concept rattaché"] else "",
    )
    for i in range(10):
        result = result.replace('\n\n\n', '\n\n')
    return result
      
for i in range(len(df)):
    s = df.iloc[i,:]
    concept_file_name = concept_to_file_name(s.Concept)
    print('{}: {}'.format(i, concept_file_name))

    with open('concepts/' + concept_file_name, 'w') as f:
        f.write(line_to_md(s))
```

```python
d = {
    "title": 'Concepts',
    "path": '/concepts/',
    "children": list('/concepts/' + df.Concept.map(lambda x: concept_to_file_name(x)[:-3]).sort_values())
}
print(json.dumps(d, indent=4))
```

```python
for definition in page.split("Voir plus"):
    _, definition = definition.split("\n    ")
    m = re.match("(.*)\n../../.... - mise à jour : ../../....\n(.*)", definition)
    titre = m.groups()[0]
    definition = m.groups()[1]
    titre_allege = replace_accents(titre
                       .replace('(', '')
                       .replace(')', '')
                       .replace('-', '-')
                       .replace('.', '-')
                       .replace(',', '-')
                       .replace('é', 'e')
                       .replace("’", '-')
                       .replace("’", '-')
                                   .replace(' ', '-')
                                   .replace('--', '-')
                                   .lower()[:78]
                      )
    print(titre_allege)
    
    m = re.search("\(([A-Z]*)\)", titre)
    if m:
        nom_fichier = m.groups()[0]  + '.md'
    else:
        nom_fichier = '_'.join(titre_allege.split()) + '.md'
    with open('glossaire/' + nom_fichier, 'w') as f:
        f.write(TEMPLATE.format(titre, definition, titre_allege))
        
    #print(definition)
    
```
